```python
import tkinter as tk
from tkinter import ttk, messagebox
import csv

# ================= FUNCIONES =================

def guardar_cliente():
    nombre = entry_nombre.get()
    telefono = entry_telefono.get()
    email = entry_email.get()
    if not nombre or not telefono:
        messagebox.showerror("Complete todos los campos obligatorios.")
        return
    with open("clientes.csv", "a", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow([nombre, telefono, email])
    messagebox.showinfo("Cliente registrado correctamente.")
    entry_nombre.delete(0, tk.END)
    entry_telefono.delete(0, tk.END)
    entry_email.delete(0, tk.END)
    mostrar_clientes()
    cargar_clientes_en_combobox()

def mostrar_clientes():
    for i in tabla_clientes.get_children():
        tabla_clientes.delete(i)
    try:
        with open("clientes.csv", newline="", encoding="utf-8") as f:
            reader = csv.reader(f)
            for fila in reader:
                tabla_clientes.insert("", tk.END, values=fila)
    except FileNotFoundError:
        pass

def guardar_servicio():
    nombre_serv = entry_servicio.get()
    precio = entry_precio.get()
    if not nombre_serv or not precio:
        messagebox.showerror("Complete todos los campos del servicio.")
        return
    try:
        float(precio)
    except ValueError:
        messagebox.showerror( "El precio debe ser un n√∫mero.")
        return
    with open("servicios.csv", "a", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow([nombre_serv, precio])
    messagebox.showinfo("Servicio registrado correctamente.")
    entry_servicio.delete(0, tk.END)
    entry_precio.delete(0, tk.END)
    mostrar_servicios()
    cargar_servicios_en_combobox()
  
  

def mostrar_servicios():
    for i in tabla_servicios.get_children():
        tabla_servicios.delete(i)
    try:
        with open("servicios.csv", newline="", encoding="utf-8") as f:
            reader = csv.reader(f)
            for fila in reader:
                tabla_servicios.insert("", tk.END, values=fila)
    except FileNotFoundError:
        pass
```
